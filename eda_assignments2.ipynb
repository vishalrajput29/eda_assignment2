```python
## IMPORTING TOOLS AND LIBRARY
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("amazon.csv")
```

```python
df
```

```python
df["rating"].dropna(inplace=True)
```

```python
df["rating"].isna().sum()
```

```python
df.dropna(axis=1,inplace = True)
```

```python
df[""].unique()
```

```python
df['rating'] = pd.to_numeric(df['rating'].str.replace('|', ''), errors='coerce').fillna(0).astype(int)
```

```python
df['rating'].mean()
```

```python
 #What is the average rating for each product category
product_category = ["product_id","product_name","product_link"]

for i in product_category:
  print(df.groupby(i)['rating'].mean().to_frame())
```

```python
# What are the top rating_count products by category
df[df["rating"]==5].category
```

```python
# What is the distribution of discounted prices vs. actual prices?
sns.scatterplot(data=df, x='discounted_price', y='actual_price')
plt.xlabel('discounted_price')
plt.ylabel('actual_price')
plt.title('Distribution between discounted_price and actual_price')
plt.grid(True)
plt.show()
```

```python
#1.Load the dataframe and ensure data quality by checking for missing values and duplicate rows. Handle missing values and remove duplicate rows if necessary.
df.isna().sum()
```

```python
#DETERMINING THE OUTLIERS IN DATA
plt.figure(figsize=(22,6))
plt.subplot(1,2,2)
sns.boxplot(data = df)
plt.title("boxplot")

plt.xticks(rotation=90)

plt.show()
```

```python
lower_cap = df["Duration (ms)"].quantile(0.05)
upper_cap = df["Duration (ms)"].quantile(0.95)
```

```python
df['Duration (ms)_capped'] = np.where(df['Duration (ms)'] < lower_cap, lower_cap,
                              np.where(df['Duration (ms)'] > upper_cap, upper_cap, df['Duration (ms)']))
```

```python
df
```

```python
plt.figure(figsize=(22,6))
plt.subplot(1,2,2)
sns.boxplot(data ="Duration (ms)_capped" )
plt.title("boxplot")
plt.show()
```
